```python

def getCriptoCot(cod):
    url = f'https://api.binance.com/api/v3/ticker/price?symbol={cod}'
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica se a resposta é bem-sucedida (código 200)
        data = response.json()
        return float(data['price'])
    except requests.exceptions.RequestException as e:
        print(f"Erro ao fazer a requisição: {e}")
        return None
    except KeyError:
        print("Chave 'price' não encontrada na resposta da API.")
        return None
```

```python
def getMoneyCot(cot, to):
    url = f'https://economia.awesomeapi.com.br/last/{cot}-{to}';
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica se a resposta é bem-sucedida (código 200)
        data = response.json()
        return float(data[f'{cot}{to}']['ask'])
    except requests.exceptions.RequestException as e:
        print(f"Erro ao fazer a requisição: {e}")
        return None
    except KeyError:
        print("Chave 'ask' não encontrada na resposta da API.")
        return None
```

```python

# URL base da API da Binance
base_url = "https://api.binance.com/api/v3"

# Endpoint para obter a lista de pares de negociação
endpoint = "/exchangeInfo"

# Faz a requisição para a API da Binance
response = requests.get(base_url + endpoint)
```

```python
simbolos = {}
# Verifica se a requisição foi bem-sucedida
if response.status_code == 200:
	# Converte a resposta para formato JSON
	data = response.json()
	
	# Obtém a lista de pares de negociação
	symbols = data["symbols"]
	
	# cria um dicionário com os pares de negociação
	for symbol in symbols:
		if symbol['status'] == 'TRADING':
			simbolos[symbol['symbol']] = [symbol['baseAsset'], symbol['quoteAsset']]
				
	
else:
	print("Falha ao obter a lista de pares de negociação.")
```

```python
lado1 = []
lado2 = []

for i in simbolos.values():
	lado1.append(i[0])
	lado2.append(i[1])
```

```python
print(set(lado1))
```

```python
for i in set(lado2):
    if i not in set(lado1):
        print(i)
```

```python
combinações = {}

for i in set(lado1):
	combinações[i] = [j[1] for j in list(simbolos.values()) if j[0] == i]
```
